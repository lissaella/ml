```python
df = pd.read_csv("/content/drive/MyDrive/breast-cancer.csv")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Summary statistics
print(df.describe())
```

```python
# Check for missing values

print(df.isnull().sum())
```

```python
# Check class distribution
print(df['diagnosis'].value_counts())
```

```python
# Visualize class distribution
sns.countplot(x='diagnosis', data=df)
plt.title('diagnosis distribution')
plt.show()
```

```python

corr = df.corr()
corr
```

```python
# Visualize the correlation matrix
plt.figure(figsize=(24, 18))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot to visualize relationships between features
sns.pairplot(df, hue='diagnosis', vars=['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean'])
plt.title('Pairplot of Features')
plt.show()
```

```python
# Visualization 6: Histogram of Features
df.hist(figsize=(18, 18), bins=20)
plt.suptitle('Histogram of Features')
plt.show()
```
